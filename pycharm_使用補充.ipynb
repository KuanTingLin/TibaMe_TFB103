```
pycharm 安裝好之後，我們要開始建立專案
打開 pycharm 會呈現如下畫面

我們可以從左側打開開過的專案
可以從右側的 open 開啟路徑
也可以透過 New Project 建立專案
```

````python
Image(filename='./resource/pycharm_project_setting/setting_project.png', width=500)
````

```
新專案需要指定專案路徑在哪(Location)
並且選擇該專案的執行環境
執行環境可以選擇新建立一個，這時需要指定我們要用什麼工具建立環境
```
```
我們也可以指定已有的環境(Existing interpreter)
這邊可以選擇我們透過 conda 建立的虛擬環境
但如果選單中沒有就點擊右方的 ...
```

````python
Image(filename='./resource/pycharm_project_setting/choose_enviroment.png', width=500)
````

```
添加已有的專案，我們選擇 conda Enviroment
並在 interpreter 放上虛擬環境的路徑(conda env list)
但記得向下找到 /bin/python(windows 會有 .exe)
若想讓其他專案也可以使用，就勾選 Make available to all project
```

# 操作設定

````python
Image(filename='./resource/pycharm_setting/setting.png', width=500)
````

```
pycharm 可以設定一些快捷鍵跟操作
從左上方的PyCharm(Mac) 或是 File(Windows)
可以找到 Preference(Mac) 或是 Settings(Windows)
```

````python
Image(filename='./resource/pycharm_setting/decrease.png', width=500)
````

```
左側有一系列的設定選項
Keymap 可以設定我們的快捷鍵操作
例如我們可以找到放大字體的快捷鍵(Increase Font Size)
也有縮小字體的快捷鍵(Decrease Font Size)
```

````python
Image(filename='./resource/pycharm_setting/keymap_setting.png', width=500)
````

```
點擊兩下我們就可以打開設定視窗
第一個是鍵盤快捷鍵
第二個是滑鼠快捷鍵
```

````python
Image(filename='./resource/pycharm_setting/setting_bar.png', width=500)
````

```
打開 mouse shortcut 之後
我們在中間灰色的窗格內進行我們想要的快捷動作，就會被記錄下來，接著只要按下 OK 就會被儲存了。
```
